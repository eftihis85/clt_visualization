```python
from scipy.stats import  rv_continuous, uniform, norm, gamma, chi, t, pareto
import matplotlib.pyplot as plt
import numpy as np

from typing import cast
```

```python
def get_run_avgs_of_dist(  dist :rv_continuous, number_of_runs: int, 
                            number_of_samples_per_run:int) -> list[float]:
    

    def get_sample_avg(n: int)-> float:   
        
        samples : list[float]= dist.rvs(size=n)
        sum_of_samples:float = cast(float, sum(samples))
        
        return cast(float, (sum_of_samples - n * dist.mean()) / ((dist.var()* n) **.5 ))

    run_avgs:list[float] = []

    for _ in range(1,number_of_runs):
        run_avgs.append(get_sample_avg(number_of_samples_per_run))


    return run_avgs
```

```python
number_of_runs = 10_000
number_of_samples_per_run = 1000
dists: list[rv_continuous] =    [   
                                    uniform(loc=0, scale=10), 
                                    gamma(a=2),
                                    chi(df=2),
                                    t(df=10),
                                    # infinite variance
                                    t(df=2),
                                    pareto(b=1.5),
                                ]


fig, ax = plt.subplots(nrows=len(dists), ncols=1, 
                       sharex=True, sharey=False, 
                       constrained_layout=True # fixes the spacing not to overlap
                       )

# fig.tight_layout() # Or equivalently,  "plt.tight_layout()"



# define norm dist params
mu, std = 0, 1

for i in range(0, len(dists)):
    dist = dists[i]
    
    ax_curr = ax[i] if len(dists) != 1 else ax
     
    avgs_of_dist = get_run_avgs_of_dist(dist=dist, 
                                        number_of_runs=number_of_runs, 
                                        number_of_samples_per_run=number_of_samples_per_run)
    
    # plot hist
    __counts, bins, _ = ax_curr.hist(avgs_of_dist, bins='auto', color='skyblue')
    
    # calc the scaled norm dist
    bin_width = bins[1] - bins[0]
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)
    pdf = norm.pdf(x, mu, std) 
    pdf_scaled =pdf * number_of_runs * bin_width
    ax_curr.plot(x, pdf_scaled, color='red', linewidth=1)
    
    # Get keyword arguments (like 'loc' and 'scale')
    kw_params = [f'{k}={v}' for k, v in dist.kwds.items()]
    
    # Combine them into one string
    all_params = ', '.join(kw_params)    
    
    dist_desc = f'{dist.dist.name} ({all_params}), var: {dist.var():,.2f}' 

    ax_curr.set_title(dist_desc)
    
plt.show()
```
